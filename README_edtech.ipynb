# Проект 1: Анализ результатов А/В-тестирования в edtech

## Описание: 
В онлайн-образовательной платформе было проведено А/В-тестирование по внедрению новой механики оплаты услуг на сайте. В ходе тестирования одной гипотезы целевой группе была предложена новая механика оплаты услуг на сайте, у контрольной группы оставалась базовая механика. Мне необходимо проанализировать итоги эксперимента и сделать вывод, стоит ли запускать новую механику оплаты на всех пользователей.  
### Задачи: 
1. Проанализировать итоги эксперимента и сделать вывод, стоит ли запускать новую механику оплаты на всех пользователей;
2. Написать 2 функции: первая будет автоматически подгружать информацию из доп. файла и на его основании пересчитывать метрики, а вторая строить графики по полученным метрикам.

**Стек**: Python (*pandas, numpy, seaborn, matplotlib, scipy.stats, requests, urllib.parse, pandahouse*), API, A/B-тест.
### Этапы работы:  
- Провела EDA, выявила аномалии (пользователи оплатили, но были неактивны) и пропущенные значения;  
- Определила и рассчитала метрики для целевой и контрольной группы (CR, AvgCheck). С помощью теста Хи-квадрат и Бутстреп (bootstrap) я проанализировала результаты A/B-теста на наличие стат значимой разницы в группах;  
- На Python реализовала функцию, которая автоматически подгружает данные из доп. файла через API и на основании дополнительных параметров пересчитывает метрики. А также функцию построения графиков по получаемым метрикам.  

## Вывод:
В результате эксперимента разница среднего чека целевой и контрольной группы была почти 35%, поэтому я приняла решение о запуске новой механики оплаты услуг, но с поправкой на улучшение конверсии, так как эта метрика не показала статзначимой разницы. 

## Описание данных
### Задача 1.
В качестве входных данных у нас есть 4 csv-файла:

**groups.csv** – файл с информацией о принадлежности пользователя к контрольной или экспериментальной группе (А – контроль, B – целевая группа)  
**groups_add.csv** – дополнительный файл с пользователями, который вам прислали спустя 2 дня после передачи данных  
**active_studs.csv** – файл с информацией о пользователях, которые зашли на платформу в дни проведения эксперимента  
**checks.csv** – файл с информацией об оплатах пользователей в дни проведения эксперимента  

### Задача 2.
Реализовать функцию, которая будет автоматически подгружать информацию из дополнительного файла **groups_add.csv** (заголовки могут отличаться) и на основании дополнительных параметров пересчитывать метрики и строить графики по получаемым метрикам.
